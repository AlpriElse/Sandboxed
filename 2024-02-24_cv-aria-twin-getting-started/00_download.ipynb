```python
datasets = list(metadata['benchmark']['dataset'].keys())
print(len(datasets))
print(datasets)
```

```python
import requests
import zipfile
import io

def download(download_url, dataset_name):
  content_length = requests.head(download_url).headers.get('Content-Length')
  if content_length is not None:
    size_in_kb = int(content_length) / 1000 / 1000 / 1000
    print(f'{dataset_name} is of size {size_in_kb}gb')

  print(f'downloading {dataset_name}')
  with requests.get(download_url, stream=True) as r:
    total_size = int(r.headers.get('content-length', 0))
    downloaded = 0

    with open(f'{dataset_name}.zip', 'wb') as f:
      for chunk in r.iter_content(chunk_size=8192):
        if chunk:
          f.write(chunk)
          downloaded += len(chunk)
          done = int(50 * downloaded / total_size)

          print(f"\r[{'=' * done}{' ' * (50-done)}] "
                f"{downloaded} / {total_size} bytes ({done * 2}%)", end='')
```
